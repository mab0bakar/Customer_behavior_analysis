```python
df [ 'Review Rating' ] = df.groupby('Category') ['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace (' ','_')
```

```python
df = df.rename ( columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# Create a column age_group 
labels = [ 'Young Adult' , 'Adult' , 'Middle_aged' , 'Senior' ]
df [ 'age_group' ] = pd.qcut (df ['age'] , q=4 , labels = labels)
```

```python
df [[ 'age','age_group',]].head(10)
```

```python
# Create column Purchae_frequency_days 
```

```python
frequency_mapping = { 
    'fortnightly' : 14,
    'weekly' : 7,
    'Monthly' :30,
    'Quarterly' :90,
    'Bi_Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90,
}
df ['purchase_frequency_days'] = df ['frequency_of_purchases'].map (frequency_mapping)
```

```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus
import pandas as pd

# Use the exact server from your SSMS connection
server = r'localhost\SQLEXPRESS01'
database = 'master' # Connecting to the one you see in SSMS

conn_str = (
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={server};"
    f"DATABASE={database};"
    f"Trusted_Connection=yes;"
    f"TrustServerCertificate=yes;"
)

params = quote_plus(conn_str)
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

try:
    with engine.connect() as conn:
        # This will list ALL databases on your server to prove it works
        df = pd.read_sql("SELECT name FROM sys.databases", conn)
        print("SUCCESS! Here are your databases:")
        print(df)
except Exception as e:
    print(f"Error: {e}")
```

```python
# 1. We specify the 'dataanalyst' database now that we know it exists
database = 'dataanalyst'

conn_str = (
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={server};"
    f"DATABASE={database};"
    f"Trusted_Connection=yes;"
    f"TrustServerCertificate=yes;"
)

params = quote_plus(conn_str)
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

# 2. Create a small table and insert data
with engine.connect() as conn:
    # Create the table
    conn.execute(sa.text("CREATE TABLE TestAnalysis (ID INT, Name VARCHAR(50), Score INT)"))
    # Insert some data
    conn.execute(sa.text("INSERT INTO TestAnalysis (ID, Name, Score) VALUES (1, 'User1', 85), (2, 'User2', 90)"))
    # Commit the changes
    conn.commit()
    print("Table created and data inserted successfully!")
```

```python
# Load the table into a Pandas DataFrame
with engine.connect() as conn:
    df = pd.read_sql("SELECT * FROM TestAnalysis", conn)

# Now you can use Pandas functions
print("--- Your Data Analyst View ---")
print(df.head())
print(f"\nAverage Score: {df['Score'].mean()}")
```
